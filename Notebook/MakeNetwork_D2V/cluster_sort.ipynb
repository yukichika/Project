```python
def suffix_generator_root(search_word,max_page,add_childs,append):
    suffix = "_" + search_word
    suffix += "_" + unicode(max_page)
    if add_childs:
        suffix += "_add_childs"
    if append:
        suffix += "_append"
    return suffix
```

```python
def suffix_generator(target=None,is_largest=False):
    suffix = ""
    if target != None:
        suffix += "_" + target
    if is_largest == True:
        suffix += "_largest"
    return suffix
```

```python
search_word = "Test"
max_page = 10
add_childs = True
append = False
save_dir = "../Data/Search"
root_dir = save_dir + suffix_generator_root(search_word,max_page,add_childs,append)

is_largest = True
target = "myexttext"
K = 10
exp_name = "K" + unicode(K) + suffix_generator(target,is_largest)
exp_dir = os.path.join(root_dir,exp_name)
nx_dir = os.path.join(exp_dir,"nx_datas")

size = 100
exp_name_new = "D" + str(size) + suffix_generator(target,is_largest)
exp_dir_new = os.path.join(root_dir,exp_name_new)
nx_dir_new = os.path.join(exp_dir_new,"nx_datas")
```

```python
"""ユークリッド距離"""
def euclid(p,q):
    weight = np.sqrt(np.power(p-q,2).sum())
    return weight
```

```python
target = 0
distance_list = []
for i in dict100.keys():
    if not i == target:
        distance = euclid(dict100[target],dict100[i])
        distance_list.append([i,distance])
distance_list
```

```python
distance_list = sorted(distance_list,key=lambda x: x[1],reverse=False)
distance_list
```

```python
distance_list_new = []
for i,list in enumerate(distance_list):
    distance_list_new.append([i,list[1]])
distance_list_new
```
