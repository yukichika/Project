```python
#nはリンク元、nbrはリンク先
for n in h:
    for nbr in G[n]:
        around_n=set(G.node[n]["from_hosts"].keys())
        around_nbr=set(G.node[n]["from_hosts"].keys())
        denom=1
        for co_host in (around_n & around_nbr):
            denom+=G.node[n]["from_hosts"][co_host]
        a[nbr]+=hlast[n]*G[n][nbr].get(weight_key,1)/denom
```
