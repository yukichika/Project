```python
import urllib.request
import urllib.error
import lxml.html
import re
from extractcontent3 import ExtractContent
```

```python
def tounicode(data):
    f = lambda d, enc: d.decode(enc)
    codecs = ['shift_jis','utf-8','euc_jp','cp932',
              'euc_jis_2004','euc_jisx0213','iso2022_jp','iso2022_jp_1',
              'iso2022_jp_2','iso2022_jp_2004','iso2022_jp_3','iso2022_jp_ext',
              'shift_jis_2004','shift_jisx0213','utf_16','utf_16_be',
              'utf_16_le','utf_7','utf_8_sig']

    for codec in codecs:
        try: return f(data, codec)
        except: continue
    return None
```

```python
def strip_tags(html):
    et = lxml.html.fromstring(html)
    try:
        title = et.find(".//title").text
    except:
        title = ""

    xpath = r'//text()[name(..)!="script"][name(..)!="style"]'
    texts = []
    for text in et.xpath(xpath):
        # if text.strip() == None:
        #     continue
        texts.append(text)

    text = ''.join(texts)
    text = re.sub("\\n","",text)
    text = re.sub("\\r","",text)
    text = re.sub("\\t","",text)
    text = re.sub("\&#13;","",text)
    return title,text
```

```python
url = "https://ja.wikipedia.org/wiki/%E3%83%A9%E3%83%BC%E3%83%A1%E3%83%B3"
```

```python
req = urllib.request.Request(url)
req
```

```python
response = urllib.request.urlopen(req)
response
```

```python
html = response.read()
html
```

```python
html_ = tounicode(html)
html_
```

```python
doc = lxml.html.fromstring(html_)
doc
```

```python
doc = lxml.html.make_links_absolute(doc,base_url=url)
doc
```

```python
title,text = strip_tags(lxml.html.tostring(doc))
```
