```python
def sort_nicely( l ):
    """ Sort the given list in the way that humans expect. """
    convert = lambda text: int(text) if text.isdigit() else text
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ]
    l.sort( key=alphanum_key )
```

```python
chasen_dir = r"../../Data/Search_Test_10_add_childs/Chasen_myexttext_largest"
files = os.listdir(chasen_dir)
files = [file for file in files if os.path.splitext(file)[1] == ".txt"]
sort_nicely(files)
corpus = []
for i,file in enumerate(files):
    with open(os.path.join(chasen_dir,file),"r") as fi:
        doc = [word.rstrip() for word in fi.readlines()]
        corpus.append(doc)
len(corpus)
```

```python
def term_to_id(term):
    if term not in vocas_id:
        voca_id = len(vocas)
        vocas_id[term] = voca_id
        vocas.append(term)
        n_v.append(0)#文書単位でカウントするため，初期化時は0
    else:
        voca_id = vocas_id[term]
    return voca_id
```

```python
def term_to_ids(terms):
    doc = []
    for term in terms:
        id = term_to_id(term)
        doc.append(id)

    """docの重複をなくして，出現語彙のカウントを+1"""
    for v in set(doc):
        n_v[v]+=1
    return doc
```
