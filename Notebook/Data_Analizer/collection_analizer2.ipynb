```python
import cPickle as pickle
from sklearn import decomposition
```

```python
doc2vecfile = "../Data/Search_Test_10_add_childs/D100_myexttext_largest/doc2vec.pkl"
with open(doc2vecfile) as fi:
    d2v = pickle.load(fi)
```

```python
d2v
```

```python
len(d2v)
```

```python
d2v.keys()
```

```python
vecs = [d2v[x] for x in d2v.keys()]
vecs
```

```python
pca = decomposition.PCA(1)
```

```python
pca.fit(vecs)
```

```python
vecs_pca = pca.transform(vecs)
vecs_pca
```

```python
len(vecs_pca)
```

```python
reg_vecs_pca = (vecs_pca-vecs_pca.min())/(vecs_pca.max()-vecs_pca.min())#0~1
reg_vecs_pca
```
