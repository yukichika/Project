```python
def suffix_generator(target=None,is_largest=False):
    suffix = ""
    if target != None:
        suffix += "_" + target
    if is_largest == True:
        suffix += "_largest"
    return suffix
```

```python
def term_to_id(term):
    if term not in vocas_id:
        voca_id = len(vocas)
        vocas_id[term] = voca_id
        vocas.append(term)
        n_v.append(0)#文書単位でカウントするため，初期化時は0
    else:
        voca_id = vocas_id[term]
    return voca_id
```

```python
def term_to_ids(terms):
    doc=[]
    for term in terms:
        id=term_to_id(term)
        doc.append(id)

    """docの重複をなくして，出現語彙のカウントを+1"""
    for v in set(doc):
        n_v[v]+=1

    return doc
```
